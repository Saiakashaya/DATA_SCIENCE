```python
#Heights mean=170, SD=10; probability >180
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 170
sd = 10
x = 180
prob = norm.sf(x, mean, sd)   # P(X > 180)
print("Probability of height > 180 cm:", prob)

x_values = np.linspace(130, 210, 500)
y = norm.pdf(x_values, mean, sd)
plt.figure(figsize=(7,5))
plt.plot(x_values, y, color="blue")

plt.fill_between(x_values, y, where=(x_values >= x),
                 color="orange", alpha=0.6, label="P(X > 180)")

plt.title("Normal Distribution of Heights")
plt.xlabel("Height (cm)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
#IQ mean=100, SD=15; probability 90–110
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 100
sigma = 15
lower = 90
upper = 110
prob = norm.cdf(upper, mu, sigma) - norm.cdf(lower, mu, sigma)
print("Probability of IQ between 90 and 110:", round(prob * 100, 2), "%")

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)
plt.figure(figsize=(8,5))
plt.plot(x, y, color="blue", linewidth=2)
plt.fill_between(x, y, where=((x >= lower) & (x <= upper)),
                 color="orange", alpha=0.6,
                 label="P(90 ≤ X ≤ 110)")
plt.title("Normal Distribution of IQ Scores")
plt.xlabel("IQ Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Weight mean=65, SD=8; probability <60
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 65
sigma = 8
X = 60
prob_less = norm.cdf(X, mu, sigma)
print("Probability of weight less than 60 kg:", round(prob_less * 100, 2), "%")

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)
plt.figure(figsize=(8,5))
plt.plot(x, y, color="blue", linewidth=2)
plt.fill_between(x, y, where=(x <= X),
                 color="orange", alpha=0.6,
                 label="P(X < 60)")
plt.title("Normal Distribution of Weights")
plt.xlabel("Weight (kg)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Exam mean=70, SD=12; probability >85
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 70
sigma = 12
X = 85
prob_greater = 1 - norm.cdf(X, mu, sigma)
print("Probability of scoring more than 85:", round(prob_greater * 100, 2), "%")
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)
plt.figure(figsize=(8,5))
plt.plot(x, y, color="blue", linewidth=2)
plt.fill_between(x, y, where=(x >= X),
                 color="orange", alpha=0.6,
                 label="P(X > 85)")
plt.title("Normal Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```
