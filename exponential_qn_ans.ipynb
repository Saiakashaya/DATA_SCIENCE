```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
mean_time = 10
t = 3
prob_within = expon.cdf(t, scale=mean_time)
print("Probability customer arrives within 3 minutes:",
      round(prob_within * 100, 2), "%")
data = np.random.exponential(scale=mean_time, size=10000)
plt.hist(data, bins=30, density=True, edgecolor='black')
plt.title("Exponential Distribution - Histogram")
plt.xlabel("Arrival Time (minutes)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```
