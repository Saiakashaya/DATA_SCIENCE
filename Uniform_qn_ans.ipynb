```python
#Lucky draw 1–100; probability 30–60
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt
a = 1
b = 100
low = 30
high = 60
dist = uniform(loc=a, scale=b-a)
probability = dist.cdf(high) - dist.cdf(low)
print("Probability of getting number between 30 and 60:",
      round(probability * 100, 2), "%")
data = np.random.uniform(1, 100, 10000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Uniform Distribution - Lucky Draw")
plt.xlabel("Numbers (1 to 100)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#Bus arrival 0–20 min; probability <5 min
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt
a = 0
b = 20
high = 5
dist = uniform(loc=a, scale=b-a)
probability = dist.cdf(high)
print("Probability bus arrives within 5 minutes:",
      round(probability * 100, 2), "%")
data = np.random.uniform(0, 20, 10000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Uniform Distribution - Bus Arrival Time")
plt.xlabel("Arrival Time (minutes)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#Game score 10–50; probability >40
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt
a = 10
b = 50
low = 40
dist = uniform(loc=a, scale=b-a)
probability = 1 - dist.cdf(low)
print("Probability of getting score greater than 40:",
      round(probability * 100, 2), "%")
data = np.random.uniform(10, 50, 10000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Uniform Distribution - Game Scores")
plt.xlabel("Score (10 to 50)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
