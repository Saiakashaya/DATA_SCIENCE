```python
import matplotlib.pyplot as plt
from scipy.stats  import binom
import numpy as np
n=10
p=0.95
k=9
prob=binom.pmf(k,n,p)
print(prob)
x = np.arange(0, n + 1)
pmf_values = binom.pmf(x, n, p)
plt.bar(x, pmf_values, color="blue", edgecolor="violet", label="Binomial PMF")

plt.bar(k, prob, color="orange", edgecolor="black", label=f"k = {k}")

plt.xlabel("Number of good items")
plt.ylabel("Probability")
plt.title(f"Binomial distribution")
plt.legend()
plt.grid(axis="y", alpha=0.3)
plt.show()
```
