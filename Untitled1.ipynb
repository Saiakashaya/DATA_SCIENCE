```python
import numpy as np
import pandas as pd
from scipy.stats import kurtosis as k

df = pd.read_csv("dataset/dataset/Balance_Sheet_final.csv")

target_column = 'Current Price'

# Convert the target column to numeric, coercing errors to NaN, then drop NaNs
df[target_column] = pd.to_numeric(df[target_column], errors='coerce')
financial_metric = df[target_column].dropna().tolist()

# Check if financial_metric is empty after cleaning
if not financial_metric:
    print(f"--- Statistics for {target_column} ---")
    print(f"No valid numeric data found for {target_column} after cleaning.")
else:
    kurto = k(financial_metric, fisher=True)
    kurto1 = k(financial_metric, fisher=False)
    mean_value = np.mean(financial_metric)
    max_value = max(financial_metric)
    min_value = min(financial_metric)

    print(f"--- Statistics for {target_column} ---")
    print(f"Excess Kurtosis (Fisher): {kurto:.4f}")
    print(f"Mean Value: {mean_value:.2f}\n")

    if kurto > 0:
        print("heavy tail more distribution")
        if (max_value - mean_value) > (mean_value - min_value):
            print("big values driving distribution higher")
        else:
            print("extreme low values present")
    elif kurto < 0:
        print("low tail less distribution and low variation")
    else:
        print("normal distribution and balanced variations")
```

```python
import pandas as pd
from scipy.stats import kurtosis

# 1. Select the first 10 columns from your dataset
subset_df = df.iloc[:, :10]

# 2. Iterate through each selected column
for col in subset_df.columns:
    # Clean data: convert to numbers and remove missing entries
    numeric_data = pd.to_numeric(subset_df[col], errors='coerce').dropna()

    # Skip columns that contain only text (like Country or Month)
    if len(numeric_data) == 0:
        print(f"Column '{col}': Text data (Skipped)")
        print("-" * 50)
        continue

    # 3. Calculate Fisher Kurtosis (Normal Distribution = 0)
    val = kurtosis(numeric_data, fisher=True)

    # 4. Classify based on the kurtosis value
    if val < 0:
        classification = "Platykurtic (Flat peak, thin tails)"
    elif val > 0:
        classification = "Leptokurtic (Sharp peak, heavy tails)"
    else:
        classification = "Mesokurtic (Normal distribution)"

    # Print the clean output
    print(f"Column:         {col}")
    print(f"Kurtosis Value: {val:.4f}")
    print(f"Distribution:   {classification}")
    print("-" * 50)
```
