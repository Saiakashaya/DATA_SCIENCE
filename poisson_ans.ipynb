```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson as p
lam=6
k=10
pos=p.pmf(k,lam)
print("probability of getting 6 visits in a minute:",pos)
x=np.arange(0,21)
fit=p.pmf(x,lam)
plt.bar(x,fit,color="pink")
plt.title("Website Visits per Minute")
plt.xlabel("no of visits in per minute")
plt.ylabel("probability")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson as p
lam=3
k=5
pos=p.pmf(k,lam)
print("probability of getting exactly 5 customers in the next 5 minutes:",pos)
x=np.arange(0,11)
plt.figure(figsize=(7,5))
fit=p.pmf(x,lam)
plt.bar(x,fit,color="purple",edgecolor="black")
plt.bar(k, pos, color="orange", edgecolor="black", label=f"k = {k}")
plt.title("Bank Customers in 5 Minutes")
plt.xlabel("no of customers")
plt.ylabel("probability")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson as p
lam=2
k=2
pos=p.pmf(k,lam)
print("Probability of getting exactly 2 cars in 30 seconds:",pos)
x=np.arange(0,8)
plt.figure(figsize=(7,5))
fit=p.pmf(x,lam)
plt.bar(x,fit,color="purple",edgecolor="black")
plt.bar(k, pos, color="orange", edgecolor="black", label=f"k = 2")
plt.title("Toll Booth Cars in 30 Seconds")
plt.xlabel("no of cars")
plt.ylabel("probability")
plt.xticks(x)
plt.legend()
plt.grid(axis="y", alpha=0.3)
plt.show()
```
