```python
import matplotlib.pyplot as plt

values = [3.1, 42.7, 55.4, 57.0]
labels = ["Study Hours", "Math", "Science", "English"]

plt.figure(figsize=(6,6))

plt.pie(values,
        labels=labels,
        autopct="%1.1f%%",
        startangle=90)

plt.title("Student Performance")
plt.show()
```

```python
import matplotlib.pyplot as plt
sizes = [40, 30, 20, 10]
labels = ['Python', 'Java', 'C++', 'Ruby']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = (0.1, 0, 0, 0)
plt.figure(figsize=(6, 6))
plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140,
    wedgeprops={'linewidth': 1.5, 'edgecolor': 'white'} # Distinct borders
)
plt.title("Programming Language Popularity")
plt.show()
```
