```python
#50 students, mean=165, SD=10; 95% CI
import numpy as np
from scipy.stats import norm

n = 50
mean = 165
sd = 10
z = 1.96
margin_error = z * (sd / np.sqrt(n))
lower_ci = mean-margin_error
upper_ci = mean+margin_error

print("Margin of Error =", round(margin_error,2))
print("\n95% Z-Confidence Interval")
print("Lower Bound =", round(lower_ci,2))
print("Upper Bound =", round(upper_ci,2))
```
