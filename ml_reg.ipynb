```python
import pandas as pd
df=pd.read_csv("Ice Cream Sales - temperatures.csv")
df.head()
```

```python
profit=df["Ice Cream Profits"]
profit
```

```python
temp=df["Temperature"]
temp
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
import numpy as np
x = np.array(df["Temperature"]).reshape(-1, 1)
y = np.array(df["Ice Cream Profits"]).reshape(-1, 1)
```

```python
print(x)
```

```python
y
```

```python
model.fit(x,y)
```

```python
new_temp=int(input("Enter your temperature:"))
new_data=np.array(new_temp).reshape(-1,1)
y_answer=model.predict(new_data)
print("Your profit is:",model.predict(new_data))
```

```python
import matplotlib.pyplot as plot
plot.plot(x,y)
plot.scatter(new_data,y_answer,color="blue")
plot.xlabel("Temperature")
plot.ylabel("Profit")
plot.title("Temperature vs Profit")
plot.show()
```
