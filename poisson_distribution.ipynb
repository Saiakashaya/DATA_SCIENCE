```python
import pandas as pd
df = pd.read_csv('Walmart_Sales.csv')
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson as p

# df is already loaded in the previous cell '2sXTLt8jsmv7' as 'Walmart_Sales.csv'
# df = pd.read_csv("Walmart.csv") # This line caused the FileNotFoundError and is redundant.

lam = int(df['Weekly_Sales'].mean())
k = np.arange(lam - 1000, lam + 1000, 100)
pos=p.pmf(k,lam)
print (pos)
x=np.arange(0,21)
fit=p.pmf(x,lam)
plt.figure(figsize=(8,5))
plt.plot(x,fit, marker='o') # Changed 'y' to 'fit'
plt.title("Poisson Distribution of Weekly Sales")
plt.xlabel("Weekly Sales")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```
