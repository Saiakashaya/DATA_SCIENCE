```python
import numpy as np

marks = np.array([78, 92, 65, 88, 95, 70])

print("Highest mark index:", np.argmax(marks))
print("Lowest mark index:", np.argmin(marks))
print("Marks above 80:", np.where(marks > 80))
print("Cumulative Sum:", np.cumsum(marks))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Month":["Jan","Feb","Mar","Apr","May"],
    "Sales":[1000,1500,1200,1800,2000]
}
df = pd.DataFrame(data)
plt.plot(df["Month"], df["Sales"])
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
import pandas as pd
data = {
    "Sample":["S1","S2","S3","S4","S5"],
    "Protein_Content":[12.5,14.2,13.8,15.1,11.9]
}
df = pd.DataFrame(data)
print(df)
print("Average Protein Content =", df["Protein_Content"].mean())
print("Maximum =", df["Protein_Content"].max())
print("Minimum =", df["Protein_Content"].min())
```
